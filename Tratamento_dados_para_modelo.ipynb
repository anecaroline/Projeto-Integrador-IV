```python
import pandas as pd
```

```python
base_dados = pd.read_csv('ocorrências_todos_estados.csv',sep=';')
base_dados.head(5)
```

```python
feriados2024 = pd.read_excel('feriados2024.xlsx')
feriados2024
```

```python
base_dados['Data'] = pd.to_datetime(base_dados['Data'])
```

```python
base_dados.shape
```

```python
novo_dataset= base_dados.copy()
```

```python
from datetime import timedelta

def classificar_data(row, feriados):
    data = row['Data']
    dia_semana = data.weekday()  # 0=segunda, 6=domingo
    hora = data.hour

    # Verificar se é véspera de feriado e se o status é "sim" ou "não"
    feriado_info = feriados[feriados['feriado'] == data.normalize()]
    if not feriado_info.empty:
        is_vespera = feriado_info['e_Vespera'].values[0] == 'Sim'
        if is_vespera:
            # Se é véspera com "sim", aplica a lógica de sexta após as 18h
            if hora >= 18:
                return "final de semana e feriado"
        else:
            # Se é véspera com "não", considera o dia todo como feriado
            return "final de semana e feriado"

    # Condições para sexta após as 18h, sábado e domingo
    if (dia_semana == 4 and hora >= 18) or dia_semana >= 5:
        return "final de semana e feriado"
    else:
        return "semana"

# Aplicar a função para criar a nova coluna
novo_dataset['categoria'] = novo_dataset.apply(classificar_data, axis=1, feriados=feriados2024)
```

Dia: Das 06:00 até as 17:59 (ou seja, das 06:00 às 18:00, excluindo a meia-noite)
Noite: Das 18:00 até as 05:59 do dia seguinte

```python
# Função para classificar o período do dia
def classificar_periodo(horario):
    if '06:00:00' <= horario < '18:00:00':
        return "dia"
    else:
        return "noite"

# Aplicar a função para criar a nova coluna
novo_dataset['Periodo do dia'] = novo_dataset['Horário'].apply(classificar_periodo)
```

```python
novo_dataset = novo_dataset[novo_dataset['Condição do tempo'] != 'Não Informado']
```

```python
novo_dataset['Condição do tempo'].value_counts(normalize=True)
```

```python
def classificar_condicao_tempo(clima):
    if clima == 'Ensolarado':
        return 'Bom'
    else:
        return 'Ruim'
novo_dataset['Condição metereologica'] = novo_dataset['Condição do tempo'].apply(classificar_condicao_tempo)

```

```python
novo_dataset['Tipo de pista'] = novo_dataset['Tipo de pista'].replace('Múltipla', 'Dupla')
```

```python
def classificar_acidente(gravidade):
    if gravidade == 'Sem Vítimas':
        return 'Não'
    else:
        return 'Sim'
    
novo_dataset['Ferido'] = novo_dataset['Gravidade do acidente'].apply(classificar_acidente)
```
